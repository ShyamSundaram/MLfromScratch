# Simple KNN
---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import math
%matplotlib inline
```

## Model
---

```python
class KNN:
    def __init__(self,X,y,k):
        self.X=X.values
        self.y=y.values
        self.k=k
        self.num_class=np.unique(y.values).shape[0] #total number of (unique) classes
        self.classes=np.unique(y.values) #the classes (unique)
        self.class_count=dict()
        for i in self.classes:
            self.class_count[i]=0  #number of points in each class is set to 0
        
    
    def predict(self,test): #predicts only one point, i.e. pass only one point to predict as argument
        if(self.X.shape[1]!=test.shape[0]):   #check if dimensions of point to classify and those of X match
            print("Number of dimensions do not match..")
            return
        l=[]
        for i in range(self.X.shape[0]):
            x1=self.X[i][0]
            x2=self.X[i][1]
            dist=math.sqrt(((test[0]-x1)**2)+((test[1]-x2)**2))     #calculate distance between the point to classify and every other point '(x,y)'
            l.append([dist, self.y[i], self.X[i][0], self.X[i][1]])  # record the [distance bewteen (x,y) ad test point, class of the point (x,y), x, y]
        l.sort()
        self.topk=l[:self.k] #select the closest k points 
        for i in range(self.k):
            self.class_count[l[i][1]]+=1 #count the number of points in topk in each class 
        return max(self.class_count,key=lambda x: self.class_count[x]) #return the class with highest count
    
    def plot(self,df,point):
        plt.scatter(df[df["y"]==0]["x1"],df[df["y"]==0]["x2"],c='orange')
        plt.scatter(df[df["y"]==1]["x1"],df[df["y"]==1]["x2"],c='green')            
        plt.scatter(point[0],point[1],c='red')
        x,y=[point[0]],[point[1]]
        for i in self.topk:
            plt.plot([i[2],point[0]],[i[3],point[1]],'y-')
        plt.title("With k="+str(self.k))
        plt.show()
```

## Generatig synthetic data
---

```python
x=np.random.randint(1,16,(30,2))
y=np.random.randint(0,2,(1,30))
ind=[i for i in range(1,31)]
df=pd.DataFrame(data=x,index=ind,columns=["x1","x2"])
df['y']=y[0].tolist()
df
```

```python
plt.scatter(df[df["y"]==0]["x1"],df[df["y"]==0]["x2"],c='orange')
plt.scatter(df[df["y"]==1]["x1"],df[df["y"]==1]["x2"],c='green')
plt.scatter(6,12,c='red')
plt.show()

X=df.drop("y",axis=1)
y=df["y"]
```

## Predicting with different k values
---

```python
k_vals=[7,11,17,23,29]
test=np.array([6,12])
for i in k_vals:
    knn=KNN(X,y,i)
    print("For K=",i)
    pred=knn.predict(test)
    c="orange"
    if(pred==1):
        c="green"
    print("Predicted class: ",knn.predict(test)," (",c,")")
    knn.plot(df,test)
    print("---------------")
```
